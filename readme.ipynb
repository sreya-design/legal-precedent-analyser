## 🧠 Key Technical Components

Multimodal PDF Parsing

Court documents are parsed using the unstructured library.

```python
from unstructured.partition.pdf import partition_pdf

elements = partition_pdf(
    strategy="hi_res",
    infer_table_structure=True,
    extract_images_in_pdf=True
)
```

Extracts:

* titles

* paragraphs

* tables

* footnotes

## Legal-Aware Chunking

Legal documents are segmented using MarkdownHeaderTextSplitter to preserve hierarchical structure.

```python
from langchain_text_splitters import MarkdownHeaderTextSplitter
```

This maintains context between legal sections and paragraphs.

## Vector Embeddings

Text chunks are converted into embeddings using OpenAI:

```python
OpenAIEmbeddings(model="text-embedding-3-small")
```

These embeddings represent legal text in a semantic vector space.

These chunks are passed to the language model as context for response generation.
